```python
import pandas as pd
import numpy as np

from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

from util_koyak import col_selector, grid_estimator
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn")
%matplotlib inline
```

```python
def conf_plot(y, y_pred):
    plt.figure(dpi=180)
    sns.heatmap(confusion_matrix(y, y_pred), cmap=plt.cm.Blues, annot=True, square=True, fmt='d', annot_kws={"size": 15});
    plt.xlabel('prediction')
    plt.ylabel('actual');
```

```python
df_genre = pd.read_csv("features_genre_6.csv", index_col=0)
df_genre.head()
```

```python
df_mfcc_cont = col_selector(['mfcc','contrast'], df_genre)
```

```python
df_mfcc_cont.head()
```

```python
X = df_mfcc_cont.drop("genre_top",axis=1)
y = df_mfcc_cont["genre_top"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.10,stratify=y, random_state=4444)
```

```python
sca = StandardScaler()
X_train_sca = sca.fit_transform(X_train)
X_test_sca = sca.transform(X_test);
```

```python
param_grid = {}
grid_estimator(SVC(), param_grid, X_train_sca, y_train, X_test_sca, y_test)
```

### Hyperparameter Tunning

```python
np.logspace(-2,1,6)
```

```python
param_grid = {'C': np.logspace(-2,1,6)}
grid_estimator(SVC(), param_grid, X_train_sca, y_train, X_test_sca, y_test)
```

```python
param_grid = {'C': [2.512], 'gamma': [0.001, 0.01, 0.1, 1, 10]}
grid_estimator(SVC(), param_grid, X_train_sca, y_train, X_test_sca, y_test)
```

### Plotting the best model

```python
svc = SVC(C=2.512)
svc.fit(X_train_sca, y_train)
y_pred = svc.predict(X_test_sca)
conf_plot(y_test, y_pred)
```

### Trying with SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
X_smoted, y_smoted = SMOTE(random_state=42).fit_sample(X_train_sca,y_train)
```

```python
svc = SVC(C=2.512)
svc.fit(X_smoted, y_smoted)
y_pred = svc.predict(X_test_sca)
conf_plot(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
accuracy_score(y_smoted, svc.predict(X_smoted))
```

### SMOTE before splitting

```python
from imblearn.over_sampling import SMOTE
```

```python
sca = StandardScaler()
X_train_sca = sca.fit_transform(X_train)
X_test_sca = sca.transform(X_test);
```

```python
svc = SVC(C=2.512)
svc.fit(X_train_sca, y_train)
y_pred = svc.predict(X_test_sca)
```

```python
conf_plot(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
accuracy_score(y_train, svc.predict(X_train_sca))
```
